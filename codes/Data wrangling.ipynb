```python
class Point():
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return 'Point({0},{1})'.format(self.x,self.y)

    def translate(self, dx, dy):
        "Add dx and dy to the point."
        self.x += dx
        self.y += dy

e = Point(1.5, 2.5)
```
